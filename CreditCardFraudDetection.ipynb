#Cleaning and changing the data for our model

```python
columns_rmv = ["Unnamed: 0", "cc_num", "first", "last", "street", "merchant","trans_num","unix_time","zip"]
```

```python
df.drop(columns_rmv, axis=1, inplace=True)
df_test.drop(columns_rmv, axis=1, inplace=True)
df.head()

```

```python
df_test.head()
```

```python
#convert the data and time
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['trans_date'] = df['trans_date_trans_time'].dt.strftime('%Y-%m-%d')
df['trans_date'] = pd.to_datetime(df['trans_date'])
df['dob'] = pd.to_datetime(df['dob'])
```

```python
#convert the data and time (test)_
df_test['trans_date_trans_time'] = pd.to_datetime(df_test['trans_date_trans_time'])
df_test['trans_date'] = df_test['trans_date_trans_time'].dt.strftime('%Y-%m-%d')
df_test['trans_date'] = pd.to_datetime(df_test['trans_date'])
df_test['dob'] = pd.to_datetime(df_test['dob'])
```

```python
#calculating age
df["age"] = df["trans_date"] - df["dob"]
```

```python
#calculating age test
df_test["age"] = df_test["trans_date"] - df_test["dob"]
```

```python
# Get Transaction Month & Year
df['trans_month'] = pd.DatetimeIndex(df['trans_date']).month
df['trans_year'] = pd.DatetimeIndex(df['trans_date']).year
```

```python
# Get Transaction Month & Year test
df_test['trans_month'] = pd.DatetimeIndex(df_test['trans_date']).month
df_test['trans_year'] = pd.DatetimeIndex(df_test['trans_date']).year
```

```python
#Calculate distance between merchant and home location
df['latitudinal_distance'] = abs(round(df['merch_lat'] - df['lat'],3))
df['longitudinal_distance'] = abs(round(df['merch_long'] - df['long'],3))
```

```python
#Calculate distance between merchant and home location test
df_test['latitudinal_distance'] = abs(round(df_test['merch_lat'] - df_test['lat'],3))
df_test['longitudinal_distance'] = abs(round(df_test['merch_long'] - df_test['long'],3))
```

```python
#drop columns after editing features
drop_columns = ['trans_date_trans_time','city','lat','long','job','dob','merch_lat','merch_long','trans_date','state']
df.drop(columns=drop_columns,inplace=True)
```

```python
df.head()
```

```python
#drop columns after editing features test
df_test.drop(columns=drop_columns,inplace=True)
```

```python
df_test.head()
```

```python
df.gender=df.gender.apply(lambda x: 1 if x=="M" else 0)
df_test.gender=df_test.gender.apply(lambda x: 1 if x=="M" else 0)
```

```python
df.head()
```

```python
df_test.head()
```

```python
#encoding of categories
df = pd.get_dummies(df, columns=['category'], prefix='category')
df_test = pd.get_dummies(df_test, columns=['category'], prefix='category')

df_test = df_test.reindex(columns=df.columns, fill_value=0)
df_test.describe()
```

```python
df.head()
```

#Train Test Split

```python
from sklearn.model_selection import train_test_split

#splitting data from train and test sets for machine learning
X_train = df.drop('is_fraud', axis=1)
y_train = df['is_fraud']
X_train['age'] = X_train['age'].dt.days
X_train['age'] = X_train['age'] / 365.25


X_test = df_test.drop('is_fraud', axis=1)
y_test = df_test['is_fraud']
X_test['age'] = X_test['age'].dt.days
X_test['age'] = X_test['age'] / 365.25
```

```python
X_train.head()
```

##Handling Data Inbalances

```python
from imblearn.over_sampling import RandomOverSampler

# Initialize the RandomOverSampler object
ros = RandomOverSampler(random_state=42)

# Apply random over-sampling to the training data
X_train_resampled, y_train_resampled = ros.fit_resample(X_train, y_train)
```

#Model Building on the data


##1. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

#Using Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

#Predictions on the test set
y_pred_rf = rf.predict(X_test)

#Classification report
report = classification_report(y_test, y_pred_rf)
print(report)
```

##2. Decision Tree


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

#sing Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

#Predictions on the test set
y_pred_dt = dt.predict(X_test)

#Classification report
report = classification_report(y_test, y_pred_dt)
print(report)
```

##3. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

#Use logistic regression
lr = LogisticRegression()
lr.fit(X_train, y_train)

#Prediction on test based upon data
y_pred_logistic = lr.predict(X_test)

#Classification report
report = classification_report(y_test, y_pred_logistic)
print(report)
```

Therefore, with an accuracy of 82% Random Forest was the best model.
